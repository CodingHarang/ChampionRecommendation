```python
import json
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
with open ('AllSoleRankDict.json', "r", encoding='utf-8') as f:
    dataDict = json.load(f)
```

```python
allDataDF = pd.DataFrame.from_dict(dataDict).transpose()
allDataDF
```

```python
allDataDF_champions = allDataDF[['wp1_champion', 'wp2_champion', 'wp3_champion', 'wp4_champion', 'wp5_champion', 'lp1_champion', 'lp2_champion', 'lp3_champion', 'lp4_champion', 'lp5_champion']]
allDataDF_champions
```

```python
dfChampions = allDataDF_champions.values.tolist()
dfChampions
```

```python
# championList에는 모든 챔피언명이 저장된다
championList = []
for i in tqdm(range(len(allDataDF))):
    if allDataDF_champions.loc[str(i), 'wp1_champion'] not in championList:
        championList.append(allDataDF_champions.loc[str(i), 'wp1_champion'])
    if allDataDF_champions.loc[str(i), 'wp2_champion'] not in championList:
        championList.append(allDataDF_champions.loc[str(i), 'wp2_champion'])
    if allDataDF_champions.loc[str(i), 'wp3_champion'] not in championList:
        championList.append(allDataDF_champions.loc[str(i), 'wp3_champion'])
    if allDataDF_champions.loc[str(i), 'wp4_champion'] not in championList:
        championList.append(allDataDF_champions.loc[str(i), 'wp4_champion'])
    if allDataDF_champions.loc[str(i), 'wp5_champion'] not in championList:
        championList.append(allDataDF_champions.loc[str(i), 'wp5_champion'])
```

```python
len(championList)
```

```python
championList.sort()
championList
```

```python
champCountDict_all = {}
champCountDict_win = {}
champCountDict_lose = {}
for i in range(161):
    champCountDict_all[championList[i]] = 0
    champCountDict_win[championList[i]] = 0
    champCountDict_lose[championList[i]] = 0
```

```python
# 단일 챔피언의 승/패 게임 수를 count한다
for i in tqdm(range(len(allDataDF))):
    champCountDict_win[allDataDF_champions.loc[str(i), 'wp1_champion']] = champCountDict_win[allDataDF_champions.loc[str(i), 'wp1_champion']] + 1
    champCountDict_win[allDataDF_champions.loc[str(i), 'wp2_champion']] = champCountDict_win[allDataDF_champions.loc[str(i), 'wp2_champion']] + 1
    champCountDict_win[allDataDF_champions.loc[str(i), 'wp3_champion']] = champCountDict_win[allDataDF_champions.loc[str(i), 'wp3_champion']] + 1
    champCountDict_win[allDataDF_champions.loc[str(i), 'wp4_champion']] = champCountDict_win[allDataDF_champions.loc[str(i), 'wp4_champion']] + 1
    champCountDict_win[allDataDF_champions.loc[str(i), 'wp5_champion']] = champCountDict_win[allDataDF_champions.loc[str(i), 'wp5_champion']] + 1
    
for i in tqdm(range(len(allDataDF))):
    champCountDict_lose[allDataDF_champions.loc[str(i), 'lp1_champion']] = champCountDict_lose[allDataDF_champions.loc[str(i), 'lp1_champion']] + 1
    champCountDict_lose[allDataDF_champions.loc[str(i), 'lp2_champion']] = champCountDict_lose[allDataDF_champions.loc[str(i), 'lp2_champion']] + 1
    champCountDict_lose[allDataDF_champions.loc[str(i), 'lp3_champion']] = champCountDict_lose[allDataDF_champions.loc[str(i), 'lp3_champion']] + 1
    champCountDict_lose[allDataDF_champions.loc[str(i), 'lp4_champion']] = champCountDict_lose[allDataDF_champions.loc[str(i), 'lp4_champion']] + 1
    champCountDict_lose[allDataDF_champions.loc[str(i), 'lp5_champion']] = champCountDict_lose[allDataDF_champions.loc[str(i), 'lp5_champion']] + 1
```

```python
for i in champCountDict_all:
    champCountDict_all[i] = champCountDict_win[i] + champCountDict_lose[i]
```

```python
champCountDict_lose
```

```python
champCountDict_all
```

```python
champWinRate = {}
for i in champCountDict_all:
    champWinRate[i] = round((champCountDict_win[i] / champCountDict_all[i]) * 100, 2)
    
champWinRate
```

```python
# 1st apriori algorithm
#threshold = 3820
freqChamp = []
for i in champCountDict_all:
    if champCountDict_all[i] > 3820:
        freqChamp.append(i)

len(freqChamp)
```

```python
# pair를 count할 때 candidate pair 생성, counting
pairCountDict_win = {}
for i in tqdm(range(len(freqChamp))):
    for j in range(i, len(freqChamp)):
        if freqChamp[i] != freqChamp[j]:
            for k in range(len(allDataDF_champions)):
                if freqChamp[i] in dfChampions[k][:5] and freqChamp[j] in dfChampions[k][:5]:
                    if freqChamp[i] + ',' + freqChamp[j] in pairCountDict_win:
                        pairCountDict_win[freqChamp[i] + ',' + freqChamp[j]] = pairCountDict_win[freqChamp[i] + ',' + freqChamp[j]] + 1
                    else:
                        pairCountDict_win[freqChamp[i] + ',' + freqChamp[j]] = 1
```

```python
pairCountDict_win
```

```python
pairCountDict_lose = {}
for i in tqdm(range(len(freqChamp))):
    for j in range(i, len(freqChamp)):
        if freqChamp[i] != freqChamp[j]:
            for k in range(len(allDataDF_champions)):
                if freqChamp[i] in dfChampions[k][5:] and freqChamp[j] in dfChampions[k][5:]:
                    if freqChamp[i] + ',' + freqChamp[j] in pairCountDict_lose:
                        pairCountDict_lose[freqChamp[i] + ',' + freqChamp[j]] = pairCountDict_lose[freqChamp[i] + ',' + freqChamp[j]] + 1
                    else:
                        pairCountDict_lose[freqChamp[i] + ',' + freqChamp[j]] = 1
```

```python
pairCountDict_lose
```

```python
for i in pairCountDict_win:
    if i not in pairCountDict_lose:
        pairCountDict_lose[i] = 0
```

```python
for i in pairCountDict_lose:
    if i not in pairCountDict_win:
        pairCountDict_win[i] = 0
```

```python
pairCountDict_all = {}
for i in pairCountDict_win:
    pairCountDict_all[i] = pairCountDict_win[i] + pairCountDict_lose[i]
```

```python
pairWinRate = {}
for i in pairCountDict_win:
    pairWinRate[i] = (round((pairCountDict_win[i] / (pairCountDict_all[i])) * 100, 2))

pairWinRate
```

```python
# candidate pair중에서 frequent pair를 선정했다
# threshold = 300
freqPairs = []
for i in pairCountDict_win:
    if pairCountDict_all[i] > 300:
        freqPairs.append(i)

len(freqPairs)
```

```python
# frequent Pair 분석
pairWinInfo = {}
rateUpPairList = []
rateDownPairList = []

for i in freqPairs:
    a = i.split(',')[0]
    b = i.split(',')[1]
    if champWinRate[a] < round((pairCountDict_win[i] / (pairCountDict_all[i])) * 100, 2):
        rateUpPairList.append(a + '/' + a + ',' + b + '/' + str(champWinRate[a]) + '/' + str(round((pairCountDict_win[i] / (pairCountDict_all[i])) * 100, 2)))
    else:
        rateDownPairList.append(a + '/' + a + ',' + b + '/' + str(champWinRate[a]) + '/' + str(round((pairCountDict_win[i] / (pairCountDict_all[i])) * 100, 2)))
        
    if champWinRate[b] < round((pairCountDict_win[i] / (pairCountDict_all[i])) * 100, 2):
        rateUpPairList.append(b + '/' + a + ',' + b + '/' + str(champWinRate[b]) + '/' + str(round((pairCountDict_win[i] / (pairCountDict_all[i])) * 100, 2)))
    else:
        rateDownPairList.append(b + '/' + a + ',' + b + '/' + str(champWinRate[b]) + '/' + str(round((pairCountDict_win[i] / (pairCountDict_all[i])) * 100, 2)))
    
    
    pairWinInfo[i] = (str(round((pairCountDict_win[i] / (pairCountDict_all[i])) * 100, 2)) + '%',
                      pairCountDict_win[i],
                      pairCountDict_lose[i],
                      a + ': ' + str(champWinRate[a]) + '%',
                      b + ': ' + str(champWinRate[b]) + '%',
                      'support: ' + str((pairCountDict_all[i]) / 61629),
                      a + '->' + b,
                      'conf: ' + str((pairCountDict_all[i]) / champCountDict_all[a]),
                      b + '->' + a,
                      'conf: ' + str((pairCountDict_all[i]) / champCountDict_all[b]),
                      'lift: ' + str((((pairCountDict_all[i]) / champCountDict_all[a]) / champCountDict_all[b])))

pairWinInfo
```

```python
# 승률이 오른 조합
for i in range(len(rateUpPairList)):
    rateUpPairList[i] = rateUpPairList[i].split('/')
rateUpPairList
```

```python
rateUpPairList.sort(key=lambda x:-float(x[3]))
rateUpPairList
```

```python
# 승률이 내려간 조합
for i in range(len(rateDownPairList)):
    rateDownPairList[i] = rateDownPairList[i].split('/')
rateDownPairList
```

```python
rateDownPairList.sort(key=lambda x:float(x[3]))
rateDownPairList
```

```python
# frequent pair들로 candidate triple들을 생성
candTrip = []
for i in tqdm(range(len(freqPairs))):
    a = freqPairs[i].split(',')[0]
    b = freqPairs[i].split(',')[1]
    for j in range(i, len(freqPairs)):
        c = freqPairs[j].split(',')[0]
        d = freqPairs[j].split(',')[1]
        if a != c and b != c:
            tmp = []
            tmp.append(a)
            tmp.append(b)
            tmp.append(c)
            tmp.sort()
            if (tmp[0] + ',' + tmp[1] + ',' + tmp[2]) not in candTrip:
                candTrip.append(tmp[0] + ',' + tmp[1] + ',' + tmp[2])
        if a != d and b != d:
            tmp = []
            tmp.append(a)
            tmp.append(b)
            tmp.append(d)
            tmp.sort()
            if (tmp[0] + ',' + tmp[1] + ',' + tmp[2]) not in candTrip:
                candTrip.append(tmp[0] + ',' + tmp[1] + ',' + tmp[2])
        if b != c and b != d:
            tmp = []
            tmp.append(b)
            tmp.append(c)
            tmp.append(d)
            tmp.sort()
            if (tmp[0] + ',' + tmp[1] + ',' + tmp[2]) not in candTrip:
                candTrip.append(tmp[0] + ',' + tmp[1] + ',' + tmp[2])

candTripDict_win = {}
candTripDict_lose = {}
for i in candTrip:
    candTripDict_win[i] = 0
    candTripDict_lose[i] = 0
```

```python
# candidate triple counting
for i in tqdm(candTripDict_win):
    for j in range(len(dfChampions)):
        a = i.split(',')[0]
        b = i.split(',')[1]
        c = i.split(',')[2]
        if a in dfChampions[j][:5] and b in dfChampions[j][:5] and c in dfChampions[j][:5]:
            candTripDict_win[i] = candTripDict_win[i] + 1
```

```python
candTripDict_all = {}
for i in candTripDict_win:
    candTripDict_all[i] = candTripDict_win[i] + candTripDict_lose[i]
```

```python
candTripDict_winRate = {}
for i in candTripDict_win:
    a = i.split(',')[0]
    b = i.split(',')[1]
    c = i.split(',')[2]
    if candTripDict_all[i] == 0:
        candTripDict_winRate[i] = -1
    else:
        candTripDict_winRate[i] = (round((candTripDict_win[i] / candTripDict_all[i]) * 100, 2),
                                   candTripDict_win[i],
                                   candTripDict_lose[i],
                                   (a + ',' + b + '->' + str(pairWinRate[a + ',' + b])),
                                   (a + ',' + c + '->' + str(pairWinRate[a + ',' + c])),
                                   (b + ',' + c + '->' + str(pairWinRate[b + ',' + c])))
        
candTripDict_winRate
```

```python
candTripDict_all = {}
for i in candTripDict_win:
    candTripDict_all[i] = candTripDict_win[i] + candTripDict_lose[i]
    
candTripDict_all
```

```python
freqTripDict = {}
for i in candTripDict_all:
    if candTripDict_all[i] > 100:
        freqTripDict[i] = candTripDict_all[i]
        
freqTripDict
```

```python
# frequent triple 분석
candTripDict_winRate = {}
rateUpTripleList = []
rateDownTripleList = []
for i in freqTripDict:
    a = i.split(',')[0]
    b = i.split(',')[1]
    c = i.split(',')[2]
    if pairWinRate[a + ',' + b] < round((candTripDict_win[i] / candTripDict_all[i]) * 100, 2):
        rateUpTripleList.append(a + ',' + b + '/' + a + ',' + b + ',' + c + '/' + str(pairWinRate[a + ',' + b]) + '/' + str(round((candTripDict_win[i] / candTripDict_all[i]) * 100, 2)))
    else:
        rateDownTripleList.append(a + ',' + b + '/' + a + ',' + b + ',' + c + '/' + str(pairWinRate[a + ',' + b]) + '/' + str(round((candTripDict_win[i] / candTripDict_all[i]) * 100, 2)))
    
    if pairWinRate[a + ',' + c] < round((candTripDict_win[i] / candTripDict_all[i]) * 100, 2):
        rateUpTripleList.append(a + ',' + c + '/' + a + ',' + b + ',' + c + '/' + str(pairWinRate[a + ',' + c]) + '/' + str(round((candTripDict_win[i] / candTripDict_all[i]) * 100, 2)))
    else:
        rateDownTripleList.append(a + ',' + c + '/' + a + ',' + b + ',' + c + '/' + str(pairWinRate[a + ',' + c]) + '/' + str(round((candTripDict_win[i] / candTripDict_all[i]) * 100, 2)))
    
        
    if pairWinRate[b + ',' + c] < round((candTripDict_win[i] / candTripDict_all[i]) * 100, 2):
        rateUpTripleList.append(b + ',' + c + '/' + a + ',' + b + ',' + c + '/' + str(pairWinRate[b + ',' + c]) + '/' + str(round((candTripDict_win[i] / candTripDict_all[i]) * 100, 2)))
    else:
        rateDownTripleList.append(b + ',' + c + '/' + a + ',' + b + ',' + c + '/' + str(pairWinRate[b + ',' + c]) + '/' + str(round((candTripDict_win[i] / candTripDict_all[i]) * 100, 2)))
    
        
    if candTripDict_all[i] == 0:
        candTripDict_winRate[i] = -1
    else:
        candTripDict_winRate[i] = (str(round((candTripDict_win[i] / candTripDict_all[i]) * 100, 2)) + '%',
                                   candTripDict_win[i],
                                   candTripDict_lose[i],
                                   (a + ',' + b + ': ' + str(pairWinRate[a + ',' + b])) + '%',
                                   (a + ',' + c + ': ' + str(pairWinRate[a + ',' + c])) + '%',
                                   (b + ',' + c + ': ' + str(pairWinRate[b + ',' + c])) + '%',
                                   a + ',' + b + '->' + c,
                                   'support: ' + str((candTripDict_win[i] + candTripDict_lose[i]) / 61629),
                                   'conf: ' + str((candTripDict_win[i] + candTripDict_lose[i]) / (pairCountDict_all[a + ',' + b])),
                                   a + ',' + c + '->' + b,
                                   'support: ' + str((candTripDict_win[i] + candTripDict_lose[i]) / 61629),
                                   'conf: ' + str((candTripDict_win[i] + candTripDict_lose[i]) / (pairCountDict_all[a + ',' + c])),
                                   b + ',' + c + '->' + a,
                                   'support: ' + str((candTripDict_win[i] + candTripDict_lose[i]) / 61629),
                                   'conf: ' + str((candTripDict_win[i] + candTripDict_lose[i]) / (pairCountDict_all[b + ',' + c])))

candTripDict_winRate
```
